```python
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("wETH_USDC_pool.csv")
```

```python
df = df.sort_values(by='blockNumber')
```

```python
df.head()
```

# 5 minute rebase

```python
ranges_5m = np.arange(df.blockNumber.min(), df.blockNumber.max(), 150)  #[0, 0.1, 0.2, 0.3 ...] 
# ranges_15m = np.arange(df.blockNumber.min(), df.blockNumber.max(), 450)  #[0, 0.1, 0.2, 0.3 ...]
# ranges_30m = np.arange(df.blockNumber.min(), df.blockNumber.max(), 900)  #[0, 0.1, 0.2, 0.3 ...]
# ranges_1hr = np.arange(df.blockNumber.min(), df.blockNumber.max(), 1800)  #[0, 0.1, 0.2, 0.3 ...] 
```

```python
labels_5m = [i for i in range(1,len(ranges_5m))]
```

```python
df['range_5m'] = pd.cut(df.blockNumber, ranges_5m, labels=labels_5m, include_lowest=True)
```

## Remove data points that are unable to group

```python
out_of_range = df.loc[df['range_5m'].isnull()]
df = df.loc[~df['range_5m'].isnull()].copy()
```

```python
# df['range_5m'] = df['range_5m'].astype('int')
```

## Aggregation

We assume Median Tick = Current Tick, hence all adjustment will be based on Median Tick

```python
df_5min_agg = df.groupby(['range_5m']).agg({
    'tick': ['min', 'median', 'max'],
    'USDC': ['sum'],
    'wETH': ['sum']
}).reset_index()

df_5min_agg.columns = ['range_5m', 'min_tick', 'median_tick', 'max_tick', 'usdc_sum', 'wETH_sum']
```

## Forward Fill

due to there are no available data points, this could mean that those ticks are remaining constant. Hence we forward fill those that empty data points to properly simulate the scenario.

```python
df_5min_agg.isnull().sum()
```

```python
df_5min_agg = df_5min_agg.fillna(method='ffill')
df_5min_agg.isnull().sum()
```

# Baseline

1. We use median ticks are the value to calculate SMA and Bollinger Bands (BB)
2. Our rolling period is 12 given that each group is 5 minutes , 12 * 5 = 60 minutes (1 hour)
3. We use 2 standard deviation (STD) spacing as our upper bound and lower bound

```python
def get_sma(df, rate):
    return df.median_tick.rolling(rate).mean()

def get_bollinger_bands(df, rate=12):
    sma = get_sma(df, rate)
    std = df.median_tick.rolling(rate).std()
#     bollinger_up = sma + std * (2 + np.where(usdc_io_ratio < 1, usdc_io_ratio, 0)) # Calculate top band
#     bollinger_down = sma - std * (2 + np.where(usdc_io_ratio > 1, usdc_io_ratio, 0)) # Calculate bottom band

    bollinger_up = np.floor(sma + std * 2) # Calculate top band
    bollinger_down = np.floor(sma - std * 2) # Calculate bottom band

    return bollinger_up, bollinger_down
```

```python
df_5min_agg['bb_max'], df_5min_agg['bb_min'] = get_bollinger_bands(df_5min_agg)
```

# Result

## Why do we need to shift the values when evaluating result

The reason is because we need to assume that when we check the status of our position, we are technically comparing our (T-1) position vs current group / 5 minute block (min and max).

```python
df_5min_agg['lag_bb_max'] = df_5min_agg['bb_max'].shift(1)
df_5min_agg['lag_bb_min'] = df_5min_agg['bb_min'].shift(1)
```

```python
result = df_5min_agg.iloc[12:]
```

```python
result['in_range'] = (result.min_tick > result.lag_bb_min) & (result.max_tick < result.lag_bb_max)
```

```python
result.head(30)
```

## Metric 1: In Range Period

```python
print(f"{np.round((result['in_range'].sum() / result.shape[0] * 100),2 )} % of the time, we are in range")
```

```python
result[['range_5m', 'min_tick', 'max_tick', 'lag_bb_max', 'lag_bb_min']].head(100).plot(figsize=(15,10))
```

## Metric 2: Tick Spacing

```python
result['upper_tick_spacing'] = result.bb_max - result.median_tick
result['lower_tick_spacing'] = result.median_tick - result.bb_min
```

```python
result[['upper_tick_spacing', 'lower_tick_spacing']] = result[['upper_tick_spacing', 'lower_tick_spacing']].apply(lambda x: np.floor(x))
```

```python
result[['upper_tick_spacing', 'lower_tick_spacing']].median()
```

# Improvements (4x Std)

```python
def get_bollinger_bands_2(df, rate=12):
    sma = get_sma(df, rate)
    std = df.median_tick.rolling(rate).std()
#     bollinger_up = sma + std * (2 + np.where(usdc_io_ratio < 1, usdc_io_ratio, 0)) # Calculate top band
#     bollinger_down = sma - std * (2 + np.where(usdc_io_ratio > 1, usdc_io_ratio, 0)) # Calculate bottom band

    bollinger_up = np.floor(sma + std * 4) # Calculate top band
    bollinger_down = np.floor(sma - std * 4) # Calculate bottom band

    return bollinger_up, bollinger_down
```

```python
df_5min_agg_2 = df.groupby(['range_5m']).agg({
    'tick': ['min', 'median', 'max'],
    'USDC': ['sum'],
    'wETH': ['sum']
}).reset_index()

df_5min_agg_2.columns = ['range_5m', 'min_tick', 'median_tick', 'max_tick', 'usdc_sum', 'wETH_sum']
```

```python
df_5min_agg_2['bb_max'], df_5min_agg_2['bb_min'] = get_bollinger_bands_2(df_5min_agg_2)
df_5min_agg_2['lag_bb_max'] = df_5min_agg_2['bb_max'].shift(1)
df_5min_agg_2['lag_bb_min'] = df_5min_agg_2['bb_min'].shift(1)
result_2 = df_5min_agg_2.iloc[12:]

result_2['in_range'] = (result_2.min_tick > result_2.lag_bb_min) & (result_2.max_tick < result_2.lag_bb_max)

print(f"{np.round((result_2['in_range'].sum() / result_2.shape[0] * 100),2 )} % of the time, we are in range")
```

```python
result_2['upper_tick_spacing'] = result_2.bb_max - result_2.median_tick
result_2['lower_tick_spacing'] = result_2.median_tick - result_2.bb_min

result_2[['upper_tick_spacing', 'lower_tick_spacing']] = result_2[['upper_tick_spacing', 'lower_tick_spacing']].apply(lambda x: np.floor(x))
result_2[['upper_tick_spacing', 'lower_tick_spacing']].median()
```

```python
result_2[['upper_tick_spacing', 'lower_tick_spacing']].mean()
```

# Improvement 3 (STD 12 on Max & Min Tick)

```python
# def get_rs(arr):
#     pos = np.where(arr > 0, arr, 0)
#     neg = np.where(arr < 0, arr, 0)
#     return np.mean(pos) / np.mean(abs(neg))


# def get_rsi(df, rate):
#     df['diff'] = df.median_tick.diff()
#     rs = df['diff'].rolling(rate).apply(get_rs)
#     rsi = round(100 - (100 / (1+rs)), 2)
#     return rsi
        
    
def get_bollinger_bands_3(df, std_rate=12):
    max_std = df.max_tick.rolling(std_rate).std()
    min_std = df.min_tick.rolling(std_rate).std()

    bollinger_up = np.floor(df.median_tick + max_std * 2) # Calculate top band
    bollinger_down = np.floor(df.median_tick - min_std * 2) # Calculate bottom band

    return bollinger_up, bollinger_down
```

```python
df_5min_agg_3 = df.groupby(['range_5m']).agg({
    'tick': ['min', 'mean', 'max'],
    'USDC': ['sum'],
    'wETH': ['sum']
}).reset_index()

df_5min_agg_3.columns = ['range_5m', 'min_tick', 'median_tick', 'max_tick', 'usdc_sum', 'wETH_sum']
```

```python
df_5min_agg_3['bb_max'], df_5min_agg_3['bb_min'] = get_bollinger_bands_3(df_5min_agg_3)
df_5min_agg_3['lag_bb_max'] = df_5min_agg_3['bb_max'].shift(1)
df_5min_agg_3['lag_bb_min'] = df_5min_agg_3['bb_min'].shift(1)
result_3 = df_5min_agg_3.iloc[12:]

result_3['in_range'] = (result_3.min_tick > result_3.lag_bb_min) & (result_3.max_tick < result_3.lag_bb_max)

print(f"{np.round((result_3['in_range'].sum() / result_3.shape[0] * 100),2 )} % of the time, we are in range")
```
